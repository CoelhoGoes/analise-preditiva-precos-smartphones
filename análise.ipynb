# 3. Testes Estatísticos

## 3.1 Teste de Normalidade - Shapiro-Wilk

```python
print("="*80)
print("TESTE DE NORMALIDADE - SHAPIRO-WILK")
print("="*80)
print("\nH0: Os dados seguem uma distribuição normal")
print("H1: Os dados NÃO seguem uma distribuição normal")
print("Nível de significância: α = 0.05\n")

# Teste para cada variável numérica
normalidade_resultados = {}

for coluna in colunas_numericas:
    # Usar amostra de até 5000 registros (limitação do teste)
    sample = df[coluna].dropna().sample(min(5000, len(df[coluna].dropna())), random_state=42)
    statistic, p_value = stats.shapiro(sample)
    normalidade_resultados[coluna] = {'statistic': statistic, 'p_value': p_value}
    
    resultado = "NORMAL" if p_value > 0.05 else "NÃO NORMAL"
    print(f"{coluna:20} | Estatística: {statistic:.4f} | p-valor: {p_value:.6f} | {resultado}")

print("\n" + "="*80)
print("Interpretação: p-valor < 0.05 → Rejeita H0 (dados não são normais)")
```

## 3.2 Teste T - Comparação de Preços por Suporte 5G

```python
print("="*80)
print("TESTE T - COMPARAÇÃO DE PREÇOS (5G vs Não-5G)")
print("="*80)
print("\nH0: Não há diferença significativa entre os preços médios")
print("H1: Há diferença significativa entre os preços médios")
print("Nível de significância: α = 0.05\n")

# Separar dados por suporte 5G
precos_5g_sim = df[df['5g_support'] == 'Yes']['price_usd']
precos_5g_nao = df[df['5g_support'] == 'No']['price_usd']

# Realizar teste T
t_stat, p_value = ttest_ind(precos_5g_sim, precos_5g_nao, equal_var=False)

print(f"Grupo com 5G:")
print(f"  N = {len(precos_5g_sim)}")
print(f"  Média = ${precos_5g_sim.mean():.2f}")
print(f"  Desvio Padrão = ${precos_5g_sim.std():.2f}\n")

print(f"Grupo sem 5G:")
print(f"  N = {len(precos_5g_nao)}")
print(f"  Média = ${precos_5g_nao.mean():.2f}")
print(f"  Desvio Padrão = ${precos_5g_nao.std():.2f}\n")

print(f"Estatística t: {t_stat:.4f}")
print(f"p-valor: {p_value:.6f}\n")

if p_value < 0.05:
    print("✓ REJEITA H0: Há diferença significativa entre os preços!")
else:
    print("✗ NÃO REJEITA H0: Não há diferença significativa entre os preços.")

print("="*80)

# Visualização
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].boxplot([precos_5g_sim, precos_5g_nao], labels=['Com 5G', 'Sem 5G'],
                patch_artist=True, boxprops=dict(facecolor='skyblue', alpha=0.7))
axes[0].set_title('Comparação de Preços - Suporte 5G', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Preço (USD)')
axes[0].grid(axis='y', alpha=0.3)

axes[1].hist(precos_5g_sim, bins=30, alpha=0.6, label='Com 5G', edgecolor='black')
axes[1].hist(precos_5g_nao, bins=30, alpha=0.6, label='Sem 5G', edgecolor='black')
axes[1].set_title('Distribuição de Preços por Suporte 5G', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Preço (USD)')
axes[1].set_ylabel('Frequência')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## 3.3 ANOVA - Comparação de Preços por Sistema Operacional (OS)

```python
print("="*80)
print("TESTE ANOVA - COMPARAÇÃO DE PREÇOS POR SISTEMA OPERACIONAL")
print("="*80)
print("\nH0: Não há diferença significativa entre as médias dos grupos")
print("H1: Pelo menos um grupo tem média significativamente diferente")
print("Nível de significância: α = 0.05\n")

# Agrupar por OS
os_groups = df.groupby('os')['price_usd'].apply(list)

# Realizar ANOVA
f_stat, p_value = f_oneway(*os_groups)

print("Estatísticas por Sistema Operacional:")
for os_name in df['os'].unique():
    precos = df[df['os'] == os_name]['price_usd']
    print(f"\n{os_name}:")
    print(f"  N = {len(precos)}")
    print(f"  Média = ${precos.mean():.2f}")
    print(f"  Desvio Padrão = ${precos.std():.2f}")

print(f"\nEstatística F: {f_stat:.4f}")
print(f"p-valor: {p_value:.6f}\n")

if p_value < 0.05:
    print("✓ REJEITA H0: Há diferença significativa entre os grupos!")
else:
    print("✗ NÃO REJEITA H0: Não há diferença significativa entre os grupos.")

print("="*80)

# Visualização
plt.figure(figsize=(12, 6))
df.boxplot(column='price_usd', by='os', patch_artist=True, figsize=(12, 6))
plt.title('Comparação de Preços por Sistema Operacional', fontsize=14, fontweight='bold')
plt.suptitle('')
plt.xlabel('Sistema Operacional')
plt.ylabel('Preço (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3.4 Teste Qui-Quadrado - Associação entre Variáveis Categóricas

```python
print("="*80)
print("TESTE QUI-QUADRADO - ASSOCIAÇÃO ENTRE OS E SUPORTE 5G")
print("="*80)
print("\nH0: As variáveis são independentes (não há associação)")
print("H1: As variáveis são dependentes (há associação)")
print("Nível de significância: α = 0.05\n")

# Criar tabela de contingência
contingency_table = pd.crosstab(df['os'], df['5g_support'])

print("Tabela de Contingência:")
print(contingency_table)
print()

# Realizar teste Qui-Quadrado
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Estatística χ²: {chi2:.4f}")
print(f"Graus de liberdade: {dof}")
print(f"p-valor: {p_value:.6f}\n")

if p_value < 0.05:
    print("✓ REJEITA H0: Há associação significativa entre OS e Suporte 5G!")
else:
    print("✗ NÃO REJEITA H0: Não há associação significativa entre as variáveis.")

print("="*80)

# Visualização
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico de barras agrupadas
contingency_table.plot(kind='bar', ax=axes[0], alpha=0.7, edgecolor='black')
axes[0].set_title('Distribuição: OS vs Suporte 5G', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sistema Operacional')
axes[0].set_ylabel('Contagem')
axes[0].legend(title='Suporte 5G')
axes[0].grid(axis='y', alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Heatmap
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu', ax=axes[1], cbar_kws={"shrink": 0.8})
axes[1].set_title('Heatmap: OS vs Suporte 5G', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Suporte 5G')
axes[1].set_ylabel('Sistema Operacional')

plt.tight_layout()
plt.show()
```
